```python
import numpy as np
from sklearn.metrics import mean_squared_error

# 1. Definir los datos
observados = np.array([3, 7, 5, 10])
predichos = np.array([2, 8, 4, 9])

# 2. Método manual (paso a paso)
# Error = (3-2)=1, (7-8)=-1, (5-4)=1, (10-9)=1
# Cuadrados = 1, 1, 1, 1
# Promedio (MSE) = 4 / 4 = 1
# Raíz (RMSE) = sqrt(1) = 1
rmse_manual = np.sqrt(np.mean((observados - predichos)**2))

# 3. Método con Scikit-Learn
mse = mean_squared_error(observados, predichos)
rmse_sklearn = np.sqrt(mse)

print(f"El RMSE calculado manualmente es: {rmse_manual}")
print(f"El RMSE usando sklearn es: {rmse_sklearn}")
```
