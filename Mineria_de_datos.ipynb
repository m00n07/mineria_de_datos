IPython Magic Commands

```python
# sirve para evaluar el deseméño de una exopresion en cuanto a tiempo
%timeit L = [n ** 2 for n in range(1000)]
```

```python
# con doble % podemos evaluar el desempeño en cuanto al tiempo de mas de una linea de codigo
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```

Understanding Data Types in Python

```python
# importa numpy y renombralo como np
import numpy as np
```

```python
# arregaglo lineal de enteros
np.array([1, 4, 2, 5, 3])
```

```python
# Al ingresar enteros y flotantes los convierte en flotantes
np.array([3.14, 4, 2, 3])
```

```python
# le estamos diciendo que sea flotante de 32 bits
np.array([1, 2, 3, 4], dtype='float32')
```

```python
np.array([1, 2, 3, 4], dtype='int32')
```

```python
# lista de comprension que genere un rango en dos argumentos el inicio y el limite
# se genera una matriz bidi
# cuando el argumento de np.array es una iteracion anidada genera una matriz
np.array([range(i, i + 3) for i in [2, 4, 6]])
```

```python
# lista de python
[range(i, i + 3) for i in [2, 4, 6]]
```
