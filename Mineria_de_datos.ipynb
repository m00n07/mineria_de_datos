#### Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

```python
# =======================================
# 4) Mapas de calor 2x2 para combinaciones
# =======================================
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)
    t[0,1] = np.count_nonzero(~a_bool & b_bool)
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool)

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')
    ax.set_title(f'Tabla 2×2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','Sí'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','Sí'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'04_heatmap_{idx}.png', dpi=180)
```

```python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a & b)
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t

tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx)
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.show()
```

Primer intento de modelar los años de escolaridad y el ingreso estimado.   
Utilizar el primer punto y el útlimo para trazar una recta:   
$$ m = \frac{y_{2}-y_{1}}{x_{2}-x_1} $$   
$$ b = y_{1} - m * x_{1} $$

```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b
linea_ingreso_estimado_new_plot
```

```python
import matplotlib.pyplot as plt

# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})')
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos
: Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.

Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error   
$$ MSE = \frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}} {n}$$   
$MSE = $ error cuadrático medio   
$n$ = número de datos  
$Y_{i}$ = valores observados   
$\hat{Y_{i}}$ = valores predichos

```python
# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```

Además del MSE (Error Cuadrático Medio), existen varios indicadores clave para evaluar modelos de regresión lineal, cada uno con un enfoque distinto sobre el rendimiento del modelo:
*   **RMSE** (Raíz del Error Cuadrático Medio): Es la raíz cuadrada del MSE. A diferencia del MSE, el RMSE expresa el error en las mismas unidades que la variable objetivo, lo que facilita su interpretación. Es útil para penalizar errores grandes, similar al MSE.
*   **MAE** (Error Absoluto Medio): Calcula el promedio de las diferencias absolutas entre las predicciones y los valores reales. Es más robusto ante valores atípicos (outliers) que el MSE, ya que no eleva al cuadrado los errores grandes.
*   $R^{2}$ (Coeficiente de Determinación): Indica qué proporción de la variación de la variable dependiente es explicada por el modelo. Va de 0 a 1; un valor más cercano a 1 indica un mejor ajuste.
*   $R^{2}$ Ajustado: Similar al $R^{2}$, pero ajusta el valor según el número de predictores (variables independientes) en el modelo. Es más preciso que el
 cuando se añaden variables innecesarias, ya que penaliza la complejidad excesiva.    
 *   **MAPE** (Error Porcentual Absoluto Medio): Mide el error en términos porcentuales, lo que facilita entender la precisión relativa del modelo (ej. "el modelo se equivoca en un 5% en promedio").

Ejercicio 4: Indicadores de Evaluación del Modelo

1. Error (Residual)
El error o residual es la diferencia entre el valor real y el valor predicho por el modelo:

$$ e_i = y_i - \hat{y}_i $$

Donde:
- $(y_i)$ es el valor real
- $( \hat{y}_i)$ es el valor predicho por el modelo
2. Error Cuadrático
El error cuadrático se obtiene elevando al cuadrado cada error individual:

$$ e_i^2 = (y_i - \hat{y}_i)^2 $$

3. Suma de Errores Cuadráticos (SSE)
La suma de errores cuadráticos es la suma de todos los errores al cuadrado:

$$ SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

4. Error Cuadrático Medio (MSE)
El Error Cuadrático Medio se obtiene dividiendo la suma de errores cuadráticos entre el número total de observaciones:
   
$$ MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

donde:
- $(n)$ es el número total de datos

```python
# Suma de errores cuadrados (SSE)
sse_sklearn = np.sum(errores_cuadrados_sklearn)
sse_sklearn
```

```python
# Error cuadrático medio (MSE)
mse_sklearn = sse_sklearn / len(anios_escolaridad)
mse_sklearn
```

```python
import pandas as pd
import numpy as np

# esto es una semilla
np.random.seed(42)
n = 200

df = pd.DataFrame({
"id_cliente": range(1, n+1), #es un id de rango de 1 hasta 200
"genero": np.random.choice(["F", "M", "No binario", "Prefiero no decir"], size=n, p=[0.45, 0.45, 0.05, 0.05]),
# metodo de numpy a seleccionar, se dice el tamaño, y la p son las probabilidades
# f es probabilidad de 45% y asi consecutivamente
"segmento": np.random.choice(["Estándar", "Oro", "Platino"], size=n, p=[0.6, 0.3, 0.1]),
# es el segmento de estandar oro platino, igual el tamaño y la probabilidades
"satisfaccion": np.random.choice([1,2,3,4,5], size=n, p=[0.1,0.2,0.35,0.25,0.1]),
# de igual manera es una semilla con el tamaño yprobabilidades
"ciudad": np.random.choice(["SLP", "CDMX", "GDL", "MTY", "QRO"], size=n, p=[0.35,0.25,0.15,0.15,0.10]),
# igual random de la lisat de los elementos
"edad": np.random.normal(35, 10, size=n).round(0).clip(18, 70).astype(int),
# se genera de una distribucionnormal, parametros media y desviacion estandar y se redondea par a valores enteros y se topa de min 18 max 70, y lo tipamos
"ingreso_mensual": np.random.lognormal(mean=9.5, sigma=0.5, size=n).round(2), # pesos MXN aprox.
# es un log normal
"temperatura_sala": np.random.normal(23, 2, size=n).round(1),
# sinulacion de temp sigue una distr normal de media, desv est que caera de 21 a 25, y se redoneda a un decimal0
"visitas_ultimo_anio": np.random.poisson(lam=3, size=n),
# es una distr de coussian, medida de lambda
"fecha_ultima_visita": pd.to_datetime("2025-12-31") - pd.to_timedelta(np.random.randint(0, 365, size=n), unit="D")
# es un dato tipo fecha pandas tiene un metodo para sumarle y restarle de metodo aleatorio
})

df.head() # nos traer los elemento de los primeros 5 renglones
```

#### Razón (media, mediana, IQR, coeficiente de variación):

```python
def resumen_razon(s):
  # regresa una serie de pandas
  # tendra el bunero de datos
    return pd.Series({
    "n": s.count(),
    "media": s.mean(),
    "mediana": s.median(),
    "sd": s.std(),
    "min": s.min(),
    "q1": s.quantile(0.25),
    "q3": s.quantile(0.75),
    "max": s.max(),
    "cv": s.std()/s.mean() if s.mean()!=0 else np.nan # one linder este equivale a realizar un condifional
    }).round(2)

razon_stats = pd.concat({
    "edad": resumen_razon(df["edad"]),
    "ingreso_mensual": resumen_razon(df["ingreso_mensual"]),
    "visitas_ultimo_anio": resumen_razon(df["visitas_ultimo_anio"]),
    "dias_desde_ultima_visita": resumen_razon(df["dias_desde_ultima_visita"])
    }, axis=1)
razon_stats
```

#### Probabilidad Condicional   
---   
La probabilidad condicional mide la probabilidad de que ocurra un evento A, sabiendo que ya ocurrió un evento B.   
Se denota así:   
$$ P(A | B) $$   
Y su formula es:   
$$ P(A | B) = \frac{P(A B)}{P(B)} $$   
donde:   


*   P(AB): probabilidad de que ocurran A y B a la vez.
*   P(B): probabilidad de que ocurra B.

### Simulación Monte Carlo de P(H|E)

```python
#@title Simulación Monte Carlo de P(H|E)
import numpy as np
import matplotlib.pyplot as plt

# Datos del problema
#            Empleado  Desempleado  Total
# Hombre        460         40       500
# Mujer         140         260      400
# Total         600         300      900

# Probabilidad analítica P(H|E) = P(H ∩ E) / P(E)
P_H_and_E = 460/900
P_E = 600/900
P_H_given_E_exact = P_H_and_E / P_E

# Simulación Monte Carlo
N = 300_000  # número de experimentos
rng = np.random.default_rng(42)

# Estados: 0=(H, E), 1=(H, D), 2=(M, E), 3=(M, D)
counts = np.array([460, 40, 140, 260], dtype=int)
probs = counts / counts.sum()

# Simulamos elegir un individuo al azar N veces
samples = rng.choice(4, size=N, p=probs)

# Evento E: empleado -> estados {0, 2}
es_empleado = np.isin(samples, [0, 2])
# Evento H ∩ E: estado 0
es_hombre_y_empleado = (samples == 0)

# Estimaciones Monte Carlo
P_H_and_E_hat = es_hombre_y_empleado.mean()
P_E_hat = es_empleado.mean()
P_H_given_E_hat = P_H_and_E_hat / P_E_hat

print(f"Valor exacto  P(H|E)     = {P_H_given_E_exact:.6f}")
print(f"Estimación MC P(H|E)     = {P_H_given_E_hat:.6f}")
print(f"P(H∩E) (MC)              = {P_H_and_E_hat:.6f}")
print(f"P(E) (MC)                = {P_E_hat:.6f}")

# Curva de convergencia de P(H|E) por prefijos
cum_H_and_E = np.cumsum(es_hombre_y_empleado)
cum_E = np.cumsum(es_empleado)
valid = cum_E > 0
P_curve = np.empty(N)
P_curve[:] = np.nan
P_curve[valid] = cum_H_and_E[valid] / cum_E[valid]

# Graficamos una muestra espaciada de puntos para eficiencia visual
idx = np.linspace(1, N, 4000, dtype=int)

plt.figure(figsize=(8, 4.8))
plt.plot(idx, P_curve[idx-1], lw=1, label='Estimación simulada')
plt.axhline(P_H_given_E_exact, color='crimson', linestyle='--', label=f"Valor exacto = {P_H_given_E_exact:.6f}")
plt.title('Convergencia de la simulación de P(H|E)')
plt.xlabel('Número de simulaciones')
plt.ylabel('Probabilidad estimada')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
